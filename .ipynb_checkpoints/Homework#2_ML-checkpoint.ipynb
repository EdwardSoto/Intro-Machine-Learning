### Solución

**(a)** Sea $\textbf{x}$ un vector como en el ejercicio, es decir, $\textbf{x}=(1,x_1,x_2)^T$ con $x_1,x_2 \in \mathbb{R}$. Note que la clasificación de este vector $\textbf{x}$ como $+1$ o $-1$ esta determinada por el signo del producto $\textbf{w}^T\textbf{x}$, donde $\textbf{w}$ esta dado como en el ejercicio y la coordenada $w_0$ determina la tolerancia dada al perceptron, es decir, las desigualdades $w_0+w_1x_1+w_2x_2>0$ y $w_0+w_1x_1+w_2x_2<0$ determinan si $\textbf{x}$ está en la región $+1$ o $-1$, respectivamente. Véase que la ecuación $w_0+w_1x_1+w_2x_2=0$ determina una línea en el plano, y lo que las desigualdades anteriores nos estan diciendo es que las regiones $+1$ y $-1$ están separadas por esta línea para cualquier $\textbf{x}$.

Ahora, si expresamos la ecuación de esta línea como $y=ax+b$, obtenemos $x_2=\frac{-w_0}{x_2}+\frac{-w_1}{w_2}x_1$ de donde podemos concluir que la pendiente es $a=\frac{-w_1}{w_2}$ y el intercepto con el eje y es $b=\frac{-w_0}{w_2}$.

**(b)** De acuerdo a nuestro razonamiento en el punto anterior tenemos que la línea que divide las dos regiones de clasificación está dada por la ecuación $y=\frac{-w_0}{w_2}+\frac{-w_1}{w_2}x$, entonces:

Para $\textbf{w}=(1,2,3)^T$, la ecuación de la línea es $y=\frac{-1}{3}-\frac{2}{3}x$ y de las desigualdades tenemos que la región de $+1$ contiene entonces a los puntos tales que $1+2x+3y>0$, es decir, $y>\frac{-1}{3}-\frac{2}{3}x$, en otras palabras, aquellos puntos por encima de la línea. Obtenemos entonces las regiones:

```python
import numpy as np  
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt 

x = np.array(range(-5,5))  
y = eval('(-1/3)+(-2/3)*x')
plt.plot(x, y, c='red')
plt.grid()
plt.fill_between(x, -3, y, color='blue',alpha=0.5)
plt.fill_between(x, y, 3, color='green',alpha=0.5)
blue_patch = mpatches.Patch(color='blue', alpha=0.5, label='-1 region')
green_patch = mpatches.Patch(color='green', alpha=0.5, label='+1 region')
line_patch = mpatches.Patch(color='red', label='Classification line')
plt.legend(handles=[line_patch, blue_patch, green_patch])
```

Para $\textbf{w}=-(1,2,3)^T$, de igual manera la ecuación de la línea es $y=\frac{-1}{3}-\frac{2}{3}x$, sin embargo, la región $+1$ esta determinada por aquellos puntos tales que $-1-2x-3y>0$, es decir, $y<\frac{-1}{3}-\frac{2}{3}x$, en otras palabras, aquellos puntos por debajo de la línea. Obtenemos entonces las regiones:

### Solución 

Para **(a)** el siguiente código resuelve el ejercicio.

```python
#Creando la funcion f objetivo, la siguiente funcion nos genera parametros reales aleatorios entre -100 y 100 
def wf(dim, mini, maxi): 
    return np.random.randint(mini*10, maxi*10, size=dim+1)/10 

#Creando el data set linealmente separable de tamaño N, para que el experimento tenga sentidos se tendera a hacer que 
#el número de muestras sea el mismo tanto para el conjunto +1 como para el conjunto -1
def gen_data_set0(N, dim ,w_f):
    validator=True
    plus1=[]
    minus1=[]
    while validator==True:
        v=np.random.randint(mini*10, maxi*10, size=dim+1)/10 
        v[0]=1
        result=np.dot(w_f,v)
        if result>0 and len(plus1)<N/2:
            plus1.append([v,1])
        if result<0 and len(minus1)<N/2:
            minus1.append([v,-1])
        if (len(plus1)+len(minus1))==N:
            validator= False
    return plus1, minus1

# Dando los parametros para la simulacion, N numero de observaciones, dim la dimension, mini el valor minimo real que toman
#los puntos maxi el valor maximo real que toman los puntos
N=20
dim=2
mini=-100
maxi=100

#Obteniendo los resultados
w_f=wf(dim, mini, maxi)
plus1, minus1 =gen_data_set0(N, dim ,w_f)

#Creando las listas de los puntos en el espacio
def esp_point(plus,minus):
    plusx1=[]
    plusy1=[]
    for i in range(len(plus1)):
        plusx1.append(plus1[i][0][1])
        plusy1.append(plus1[i][0][2])
    minusx1=[]
    minusy1=[]
    for i in range(len(minus1)):
        minusx1.append(minus1[i][0][1])
        minusy1.append(minus1[i][0][2])

    return plusx1, plusy1, minusx1, minusy1
    
plusx1, plusy1, minusx1, minusy1=esp_point(plus1,minus1)
    
#Ploteando los resultados 
x= np.linspace(-100,100,100)
y=-(w_f[0]/w_f[2])-(w_f[1]/w_f[2])*x

plt.ylim(-100,100)
plt.xlim(-100,100)
plt.plot(plusx1,plusy1, marker='+', markersize='8' , linewidth=0, label="+1 points")
plt.plot(minusx1,minusy1, marker='$-$', markersize='8', linewidth=0, label="$-$1 points")
plt.plot(x,y,'r', label="Objective function $f$")
plt.title("Objective function and data set")
plt.xlabel("x label")
plt.ylabel("y label")
plt.legend(bbox_to_anchor=(1.0, 1.0))
plt.grid()
```

Para **(b)** se implemente el algoritmo del perceptron en el siguiente código, con los mismos parámetros de simulación de **(a)**. Note que al final que $g$ no es muy parecida a $f$.

```python
#Perceptron algoritmo, ds es la suma de los puntos etiquitados como +1 y -1. 
def perceptron(ds,dim):
    count=0
    w_g=np.zeros(dim+1)
    validator=True
    while validator==True:
        error=0
        for i in range(len(ds)):
            v=ds[i]
            if (np.sign(np.dot(w_g,v[0])))!=(np.sign(v[-1])):
                error+=1
                count+=1
                for j in range(len(w_g)):
                    w_g[j]=w_g[j]+v[-1]*v[0][j]
        if error==0:
            validator=False
        
    return w_g, count    

#Corriendo el algoritmo
ds=plus1+minus1
w_g, count= perceptron(ds,dim)

#ploteando los resultados
y1=-(w_g[0]/w_g[2])-(w_g[1]/w_g[2])*x

plt.ylim(-100,100)
plt.xlim(-100,100)
plt.plot(plusx1,plusy1, marker='+', markersize='8' , linewidth=0, label="+1 points")
plt.plot(minusx1,minusy1, marker='$-$', markersize='8', linewidth=0, label="$-$1 points")
plt.plot(x,y,'r', label="Objective function $f$")
plt.plot(x,y1, 'g', label='Obtained perceptron function $g$')
plt.title("Perceptron output")
plt.xlabel("x label")
plt.ylabel("y label")
plt.legend(bbox_to_anchor=(1.0, 1.0))
plt.grid()
print('Objective function:', w_f, ', perceptron function:', w_g, ', updates:', count )
```

Para **(c)** simplemente generamos un nuevo data set con los mismos parámetros anteriores. Note que de igual manera $g$ no es muy parecida a $f$, de echo $g$ se parece menos a $f$. Por otro lado, para este data set se tomó más actualizaciones que en para el data set de **(b)**.

```python
#Generando un nuevo Data set 
N=20
dim=2
plus1, minus1=gen_data_set0(N, dim ,w_f)

#Corriendo el algoritmo
ds=plus1+minus1
w_g, count= perceptron(ds,dim)

#Creando las listas de los puntos en el espacio
plusx1, plusy1, minusx1, minusy1=esp_point(plus1,minus1)

#Ploteando los resultados
x= np.linspace(-100,100,100)
y=-(w_f[0]/w_f[2])-(w_f[1]/w_f[2])*x
y1=-(w_g[0]/w_g[2])-(w_g[1]/w_g[2])*x

plt.ylim(-100,100)
plt.xlim(-100,100)
plt.plot(plusx1,plusy1, marker='+', markersize='8' , linewidth=0, label="+1 points")
plt.plot(minusx1,minusy1, marker='$-$', markersize='8', linewidth=0, label="$-$1 points")
plt.plot(x,y,'r', label="Objective function $f$")
plt.plot(x,y1, 'g', label='Obtained perceptron function $g$')
plt.title("Perceptron output")
plt.xlabel("x label")
plt.ylabel("y label")
plt.legend(bbox_to_anchor=(1.0, 1.0))
plt.grid()
print('Objective function:', w_f, ', perceptron function:', w_g, ', updates:', count )
```

Para **(g)** manipulamos nuestro codigo del perceptron para que ahora tome los ejemplos de entrenamiento de manera aleatoria. Creamos una lista con los resultados obtenidos teniendo como entrada la misma funcion objetivo y data set de **(f)**.

```python
#Modificamos el algoritmo del perceptron para tomar puntos aleatorios en nuestro entrenamiento 
def perceptron_random(ds,dim):
    count=0
    w_g=np.zeros(dim+1)
    validator=True
    while validator==True:
        error=0
        random=np.random.randint(0, 999, size=1000)
        for i in range(len(random)):
            v=ds[random[i]]
            if (np.sign(np.dot(w_g,v[0])))!=(np.sign(v[-1])):
                count+=1
                for j in range(len(w_g)):
                    w_g[j]=w_g[j]+v[-1]*v[0][j]
        for i in range(len(ds)):
            v=ds[i]
            if (np.sign(np.dot(w_g,v[0])))!=(np.sign(v[-1])):
                error+=1
                break
        if error==0:
            validator=False
        
    return w_g, count  

#Corriendo el algoritmo 100 veces
act_list=[]
for i in range(100):
    w_g, count= perceptron_random(ds,dim)
    act_list.append(count)
```

### Solución
**(a)** Sea $t$ una variable aleatoria finita, $\alpha$ una constante positiva y $s$ un parámetro positivo. Definamos la siguiente variable aleatoria indicadora: 
$$
I(t\geq \alpha)=
\begin{cases}
1&\text{si $t\geq \alpha$}\\
0&\text{en otro caso}\\
\end{cases}
$$
Ahora, es claro que al ser $e^x$ una función monótona creciente, $e^{st}$ será entonces una función monótona creciente dependiente de $t$ (teniendo a 0 como una asíntota horizontal). Note lo siguiente: <br> <br>
$\cdot$ Si $t< \alpha$ entonces $e^{s \alpha}I(t\geq \alpha)=e^{s \alpha}\cdot 0=0 \leq e^{st}$. <br>
$\cdot$ Si $t \geq \alpha$ entonces $s\alpha \leq st \Rightarrow e^{s\alpha} \leq e^{st}=e^{s\alpha}\cdot 1 \leq e^{st}=e^{s\alpha}I(t\geq \alpha) \leq e^{st}$ <br><br>
Podemos concluir entonces que $e^{s\alpha}I(t\geq \alpha) \leq e^{st}$. Como $e^{st}$ es monótona creciente, al tomar valor esperado se mantiene la desigualdad a ambos lados, además como $e^{s\alpha}$ es un número (positivo) y por propiedades del valor esperado se tiene: <br> <br>
$$
\begin{align*}
& E[e^{s\alpha}I(t\geq\alpha)] \leq E[e^{st}] \\
\Rightarrow & E[e^{s\alpha}] E[I(t\geq\alpha)] \leq E[e^{st}] \\
\Rightarrow & e^{s\alpha} E[I(t\geq\alpha)] \leq E[e^{st}]
\end{align*}
$$ <br>
Note ahora que $E[I(t\geq\alpha)]=0\cdot P(t<\alpha)+1\cdot P(t\geq \alpha)=P(t\geq \alpha)$. Reemplazando llegamos a lo que se quería:
<br> <br>
$$
\begin{align*}
& e^{s\alpha} E[I(t\geq\alpha)] \leq E[e^{st}]\\
\Rightarrow & e^{s\alpha} P(t\geq \alpha) \leq E[e^{st}]\\
\Rightarrow &  P(t\geq \alpha) \leq e^{-s\alpha} E[e^{st}]\\
\Rightarrow &  P(t\geq \alpha) \leq e^{-s\alpha} T(s)
\end{align*}
$$

**(b)** Sean $u_1,u_2,...,u_N$ $N$ variables aleatorias independientes e idénticamente distribuidas, $u=\frac{1}{N}\sum_{n=1}^N u_n$ y $U(s)=E_{u_n}(e^su_n)$ para toda $u_n$. De la parte **(a)** tomemos $t=Nu$ y $\alpha=N\alpha$, entonces:<br><br>
$$
\begin{align*}
& P(Nu \geq N \alpha) \leq e^{-sN \alpha} E[e^{sNu}] \\
\Rightarrow &  P(u \geq \alpha) \leq e^{-sN \alpha} E \left[e^{sN\frac{1}{N}\sum \limits_{n=1}^N u_n} \right] \\
\Rightarrow &  P(u \geq \alpha) \leq e^{-sN \alpha} E\left[ \prod_{n=1}^N e^{s u_n} \right]
\end{align*}
$$
<br>
Como las $u_i$ son independientes e idénticamente distribuidas podemos sacar la productoria del valor esperado y dejar todo en términos de una sola variable, obtenemos lo que queríamos:
$$
\begin{align*}
&  P(u \geq \alpha) \leq e^{-sN \alpha} E\left[ \prod_{n=1}^N e^{s u_n} \right] \\
\Rightarrow &  P(u \geq \alpha) \leq e^{-sN \alpha} \prod_{n=1}^NE\left[ e^{s u_n} \right] \\
\Rightarrow &  P(u \geq \alpha) \leq e^{-sN \alpha} \left(E[ e^{s u_n}]\right)^N  \\
\Rightarrow &  P(u \geq \alpha) \leq \left( e^{-s\alpha} E[ e^{s u_n}]\right)^N \\
\Rightarrow &  P(u \geq \alpha) \leq \left( e^{-s\alpha} U(s)\right)^N
\end{align*}
$$

**(c)** Supongamos que $P(u_n=0)=P(u_n=1)=\frac{1}{2}$, por la definicion del valor esperado de una funcion de una variable aleatoria tenemos entonces que: <br>
$$U(s)=E[e^{su_n}]=e^{s\cdot0}P(u_n=0)+e^{s\cdot1}P(u_n=1)=1\cdot\frac{1}{2}+e^s\cdot\frac{1}{2}=\frac{1}{2}(e^s+1)$$
Note entonces que $U(s)$ queda en funcion de $s$. Sea $\alpha$ fijo, $0<\alpha<1$, para minimizar $e^{-s\alpha}U(s)=e^{-s\alpha}\frac{1}{2}(e^s+1)$ calcularemos su derivada respecto de $s$ y igualaremos a 0 para hayar un minimo. Sea $g(s)=e^{-s\alpha}\frac{1}{2}(e^s+1)=\frac{1}{2}(e^{s(1-\alpha)}+e^{-s\alpha})$, entonces:
$$g'(s)= \frac{1}{2}(e^{s(1-\alpha)}(1-\alpha)-\alpha e^{-s\alpha})$$
Al hacer $g'(s)=0$, se llega a:
$$
\begin{align*}
&e^{s(1-\alpha)}(1-\alpha)-\alpha e^{-s\alpha}=0 \\
\Rightarrow & ln(e^{s(1-\alpha)}(1-\alpha)) = ln(\alpha e^{-s\alpha}) \\
\Rightarrow & s(1-\alpha)+ln(1-\alpha)= ln(\alpha)-s\alpha \\
\Rightarrow & s(1-\alpha)+s\alpha= ln\left(\frac{\alpha}{1-\alpha}\right)\\
\Rightarrow & s= ln\left(\frac{\alpha}{1-\alpha}\right)
\end{align*}
$$
Por tanto, $s=ln\left(\frac{\alpha}{1-\alpha}\right)$ es el minimo buscado.

**(d)** Veamos primero que $E[u]=\frac{1}{2}$. 
$$
\begin{align*}
E[u] &= E\left(\frac{1}{N} \sum_{i=1}^N u_i \right) \\
& = E\left(\frac{1}{N} \right)\cdot E \left( \sum_{i=1}^N u_i \right) \\
& = \frac{1}{N}\cdot  \sum_{i=1}^N E \left( u_i \right) \\
& = \frac{1}{N}\cdot  \sum_{i=1}^N \left(0 P(u_i=0)+1P(u_i=1) \right) \\
& = \frac{1}{N}\cdot  \sum_{i=1}^N \frac{1}{2} \\
& = \frac{1}{2}
\end{align*}
$$
<br>Ahora, por **(b)** y **(c)** sabemos que la desigualdad $P(u \geq \alpha) \leq \left( e^{-s\alpha} U(s)\right)^N=\left( e^{-s\alpha} \frac{1}{2}(e^s+1) \right)^N$, se cumplirá incluso para $s= ln\left(\frac{\alpha}{1-\alpha}\right)$. <br>
Manipulemos primero a $e^{-s\alpha} \frac{1}{2}(e^s+1)$ para llegar a nuestro objetivo, note que: <br><br>
$$
\begin{align*}
e^{-s\alpha} \frac{1}{2}(e^s+1) &= \frac{1}{2} \left( e^{ln\left(\left(\frac{\alpha}{1-\alpha}\right)^{-\alpha}\right)} \cdot \left( e^{ln \left(\frac{\alpha}{1-\alpha}\right)} +1 \right) \right) \\
&= \frac{1}{2} \left( \left(\frac{\alpha}{1-\alpha}\right)^{-\alpha} + \left(\frac{\alpha}{1-\alpha}\right)^{-\alpha+1}\right)\\
&= \frac{1}{2}\left(\frac{\alpha^{-\alpha}\left(1-\alpha \right)}{\left(1-\alpha \right)^{-\alpha+1}} + \frac{\alpha^{-\alpha+1}}{\left(1-\alpha \right)^{-\alpha+1}}\right) \\
& = \frac{1}{2} \left(\frac{1}{\alpha^{\alpha}\cdot(1-\alpha)^{(1-\alpha)}}  \right) \\
& = \frac{1}{2} \cdot\frac{1}{2^{\displaystyle log_2 \left(\alpha^{\alpha}\cdot(1-\alpha)^{(1-\alpha)} \right)}} \\
&= \frac{1}{2^{\displaystyle1+log_2 \left(\alpha^{\alpha}\cdot(1-\alpha)^{(1-\alpha)} \right)}} \\
&= 2^{\displaystyle-\left(1+log_2 \left(\alpha^{\alpha}\cdot(1-\alpha)^{(1-\alpha)} \right) \right)} \\
&= 2^{\displaystyle -\left(1+ \alpha log_2(\alpha) +(1-\alpha)log_2(1-\alpha) \right)} 
\end{align*} 
$$
<br>
Si tomamos en la desigualdad $\alpha =E[u]+ \varepsilon= \frac{1}{2} + \varepsilon$, para $0<\varepsilon < \frac{1}{2}$. Por nuestro razonamiento anterior: <br><br>
$$
P(u \geq E[u]+ \varepsilon) \leq  2^{-\beta N}
$$
<br>Donde $\beta=1+ \left(\frac{1}{2} + \varepsilon \right) log_2 \left(\frac{1}{2} + \varepsilon \right) + \left(\frac{1}{2} - \varepsilon \right)log_2 \left(\frac{1}{2} - \varepsilon \right)$. Para ver que $\beta > 0$, veamos a $\beta$ como función de $\varepsilon$. Derivemos a $\beta$ respecto de $\varepsilon$, veamos que su derivada es siempre positiva y además tiene a 0 como un punto critico. <br><br>
$$
\begin{align*}
\beta' & = log_2\left(\frac{1}{2}+\varepsilon\right)+\left(\frac{1}{2}+\varepsilon \right)\frac{2}{ln(2)(1+2\varepsilon)}-log_2\left(\frac{1}{2}-\varepsilon \right)-\left(\frac{1}{2}-\varepsilon \right)\frac{2}{ln(2)(1-2\varepsilon)} \\
&= log_2\left(\frac{1}{2}+\varepsilon\right)-log_2\left(\frac{1}{2}-\varepsilon \right) \\
&= log_2 \left( \frac{\frac{1}{2}+\varepsilon}{\frac{1}{2}-\varepsilon}\right) >0
\end{align*} 
$$
<br> Notese que esta derivada se hace 0 cuando $\varepsilon=0$, esto nos permite decir que para $\varepsilon>0$, $\beta$ es positiva y además creciente. Por tanto, la cota disminuye exponencialmente a medida que $N$ y $\beta$ crecen.
